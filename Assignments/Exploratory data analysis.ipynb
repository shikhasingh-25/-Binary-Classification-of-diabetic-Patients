## Removing outliers

```python
def drop_by_std(dataf, factor):
    
    lower =dataf['trestbps'].mean() - dataf['trestbps'].std() * factor
    upper =dataf['trestbps'].mean() + dataf['trestbps'].std() * factor   
    print(f'Lower: {lower}', f'Upper: {upper}')
    output=dataf[(dataf['trestbps'] < upper) & (dataf['trestbps'] > lower)]
    return output

df = drop_by_std(final_data, 2)
df
```

```python
plt.boxplot(df['trestbps'])
```

```python
def drop_by_std(dataf, factor):
    
    lower =dataf['chol'].mean() - dataf['chol'].std() * factor
    upper =dataf['chol'].mean() + dataf['chol'].std() * factor   
    print(f'Lower: {lower}', f'Upper: {upper}')
    output=dataf[(dataf['chol'] < upper) & (dataf['chol'] > lower)]
    return output

data_frame = drop_by_std(df, 2)
data_frame
```

```python
plt.boxplot(data_frame['chol'])
```

## Corelation Matrix

```python
corr_matrix = data_frame.corr()
fig, ax = plt.subplots(figsize=(12, 12))
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                cmap="Greens");
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```
