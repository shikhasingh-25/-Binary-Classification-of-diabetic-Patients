## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
#sine wave using matplotlib
x = np.linspace(1,10,1000)
plt.plot(x,np.sin(x),color='black',linestyle='-.')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sine Wave")
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
store=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
Clothing=np.array([82,78,48,46,39])
Equipment=np.array([48,31,51,29,18])
Accessories=np.array([15,20,14,13,9])
var=[x for x, _ in enumerate(store)]
plt.bar(var,Clothing,width=0.7,label='Clothing',color='green',bottom=Equipment+Accessories)
plt.bar(var,Equipment,width=0.7,label='Equipments',color='red',bottom=Accessories)
plt.bar(var,Accessories,width=0.7,label='Accessories',color='blue')
plt.xticks(var,store,rotation=90)
plt.xlabel("Cities")
plt.ylabel("Stores")
plt.title("Stores in cities")
plt.grid(False)
plt.legend(loc='upper right')
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
data=sns.load_dataset('titanic')
```

```python
data.head()
```

```python
data.tail()
```

```python
#Bar graph
titan=sns.countplot(x='pclass',hue='survived',palette='spring',data=data)
titan.set(title='Survival of passenger against Passenger class')
plt.legend(loc="upper left")
plt.show()
```

## Q.10 For Tips Dataset
- Load the dataset using seaborn
- Draw the Strip Plot between 'total bill' and 'day'
- Rotate xticks with an angle of 45 Degree

```python
#Loading the data
tips=sns.load_dataset('tips')
```

```python
#Strip plot
sns.stripplot(x='day',y='total_bill',data=tips,jitter=1)
plt.title('Variation of total bill with days')
plt.xticks(rotation=45)
```
